#### Does higher price lead to higher reviews?

```
import matplotlib.pyplot as plt

#plot price against points
  
y = df['points']
x = df['price']
  
plt.plot(x, y)  # Plot the chart
plt.show()  # display
```

#### This plot suggests no. 
#### Wine price within a range of 0 to 500 dollars has a wide range of points, including the largest amount of high reviews. After 500 dollars, the points plateau. Wine priced more than 3000 dollars has less points than wine priced below 500.
